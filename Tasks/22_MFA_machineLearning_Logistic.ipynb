```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
#x
digits.data.shape
X = digits.data
```

```python
#y
digits.target.shape
y = digits.target
```

```python
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:10],digits.target[0:10])):
    plt.subplot(1,10,index + 1)
    plt.imshow(np.reshape(image , (8,8)), cmap=plt.cm.gray)
    plt.title('training: %i\n' % label ,fontsize = 20)
```

```python
# [0:5] decide the number of picture if we change it.
#it will convert the number of file from 5 to 10
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:5],digits.target[0:5])):
    plt.subplot(1,10,index + 1)
    plt.imshow(np.reshape(image , (8,8)), cmap=plt.cm.gray)
    plt.title('training: %i\n' % label ,fontsize = 20)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
print("x_train input data: ", x_train.shape)
print("x_test input data: ", x_test.shape)
print("y_train input data: ", y_train.shape)
print("y_test input data: ", y_test.shape)
```

```python
#model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(x_train,y_train)
model
```

```python
#predection
model.predict(x_test[0:20])
predections = model.predict(x_test)
predections
```

```python
# accuracy test
score = model.score(x_test,y_test)
print("the accuracy score is: "), score
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,predections)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot= True, fmt= " .3f", linewidths=.5, square = True,cmap = 'Spectral');
plt.ylabel('Actual Output');
plt.ylabel('Predected Output');
all_sample_title = 'Accuracy Score : {0}'.format(score)
plt.title(all_sample_title, size = 35);
```

```python
print(cm)
```

```python
# Getting miss classification label

import numpy as np
import matplotlib.pyplot as plt
index = 0
misclassifiedIndexes = []
for label, predict in zip(y_test,predections):
    if label != predict:
        misclassifiedIndexes.append(index)
        index +=1

```

```python
# plotting missclassified labels with known table
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
for plotIndex, badIndex in enumerate(misclassifiedIndexes[5:10]):
    plt.subplot(1,5,plotIndex + 1)
plt.imshow(np.reshape(x_test[badIndex],(8,8)), cmap = plt.cm.gray)
plt.title("Predicted: {}, Actual: {} ".format(predections[badIndex], y_test[badIndex]), fontsize = 15)
```
